```python
import random
import pandas as pd

[w1, w2, w3, w4, w5] = [0.242, 0.244, 0.244, 0.243, 0.245]

vals = []

for i in range(10000):
    x1 = random.randint(30, 100)
    x2 = random.randint(30, 100)
    x3 = random.randint(30, 100)
    x4 = random.randint(30, 100)
    x5 = random.randint(30, 100)

    eq = w1 * x1 + w2 * x2 + w3 * x3 + w4 * x4 + w5 * x5

    vals.append([x1, x2, x3, x4, x5, eq])

df = pd.DataFrame(vals, columns=['no of sites', 'inc mat', 'freq breakdown', 'inacc cost', 'satisfaction','delay rate'])

df.to_csv('data.csv', index=False)

df.head(1000)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

df = pd.read_csv('data.csv')

x = df[['no of sites', 'inc mat', 'freq breakdown', 'inacc cost', 'satisfaction']]
y = df['delay rate']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

mdl = RandomForestRegressor(n_estimators=100, max_depth=6, random_state=42)
mdl.fit(x_train, y_train)

pred = mdl.predict([[66, 66, 66, 66, 66]])
print("Predicted value (RFR):", pred[0])

accuracy = mdl.score(x_train[:100], y_train[:100]) * 100
print("Accuracy (RFR):", accuracy)

print("\n")

mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)

print("\n")
print("Mean Square Error of RandomForest Regression:", mse)
print("R2 score of RandomForest Regression:", r2)
print("Mean Absolute Error of RandomForest Regression:", mae)

plt.scatter(x_train['no of sites'], y_train, color='b', label='Actual')
plt.plot(x_train['no of sites'], mdl.predict(x_train), color='black', linewidth=3, label='Predicted')
plt.xlabel('no of sites')
plt.ylabel('delay rate')
plt.title('RandomForestRegressor: no of sites vs. delay rate')
plt.legend()
plt.grid(True)
plt.show()
```
